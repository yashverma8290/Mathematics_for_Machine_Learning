# Practical Use of Central Limit Theorem (CLT) in AI & ML

In Machine Learning (ML) and Artificial Intelligence (AI), the Central Limit Theorem (CLT) is essential because many statistical models assume normality, and CLT helps us approximate normality even when the raw data is not normally distributed.

# Real-World Use Case: CLT in Predicting Customer Spending

Imagine an e-commerce company wants to predict the average customer spending per day. Individual customer spending is highly variable and not normally distributed (some customers spend very little, and some spend a lot). But thanks to CLT, if we take many samples of daily spending averages, the distribution of those averages becomes normal, allowing us to apply ML models and statistical tests effectively.

# Python Code: Simulating CLT in Customer Spending Data
We will:
1) Generate a non-normal distribution (customer spending data).
2) Take multiple random samples and compute their means.
3) Show how the sample means form a normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Generate Non-Normal Data (Right-Skewed Customer Spending)
np.random.seed(42)
customer_spending = np.random.exponential(scale=50, size=10000)  # Skewed spending data

# Step 2: Take Many Random Samples and Compute Their Means
sample_means = []
sample_size = 50  # Taking 50 samples at a time
num_samples = 1000  # Number of samples to take

for _ in range(num_samples):
    sample = np.random.choice(customer_spending, size=sample_size)
    sample_means.append(np.mean(sample))

# Step 3: Plot the Original Skewed Distribution vs. Sample Means Distribution
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Original Customer Spending Distribution (Highly Skewed)
sns.histplot(customer_spending, bins=50, kde=True, color='red', ax=axes[0])
axes[0].set_title("Original Customer Spending (Skewed)")
axes[0].set_xlabel("Spending Amount")
axes[0].set_ylabel("Frequency")

# Distribution of Sample Means (Approximately Normal)
sns.histplot(sample_means, bins=30, kde=True, color='blue', ax=axes[1])
axes[1].set_title("Distribution of Sample Means (Normal via CLT)")
axes[1].set_xlabel("Mean Spending per Sample")
axes[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```
